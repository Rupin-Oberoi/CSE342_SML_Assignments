# Q1 (a)

```python
mean1, mean2 = 2, 5
sd1, sd2 = 1, 1
dist1 = stats.norm(mean1, sd1)
dist2 = stats.norm(mean2, sd2)
s = np.linspace(-50, 50, 10000)
plt.plot(s, dist1.pdf(s))
plt.xlim(-15,15)
plt.xlabel("x")
plt.ylabel("p(x|w1)")
plt.show
```

```python
s = np.linspace(-50, 50, 10000)
plt.plot(s, dist2.pdf(s))
plt.xlim(-15,15)
plt.xlabel("x")
plt.ylabel("p(x|w2)")
plt.show
```

```python
s = np.linspace(-50, 50, 10000)
np.seterr(divide='ignore', invalid='ignore')
plt.plot(s,dist1.pdf(s)/dist2.pdf(s))
plt.xlim(-40,40)
plt.xlabel("x")
plt.ylabel("p(x|w1)/p(x|w2)")
plt.show
```
